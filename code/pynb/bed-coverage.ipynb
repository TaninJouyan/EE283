```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
columnName=['start','end','cov']
a4bg= pd.read_csv('../../data/processed/DNAseq/bams/A4.bg' , header=None, sep='\t', names=columnName)
a4bg
```

```python
a4bg['mean-nt']=a4bg[['start','end']].mean(axis=1)
a4bg
coverageplot=plt.plot(a4bg['mean-nt'],a4bg['cov'] ,linestyle='-', color='purple')
plt.xlabel("Nucleotide")
plt.ylabel("Coverage")
plt.title("Coverage of chromosome X")
plt.grid(True)
plt.savefig("coverageplot.png")
plt.show()
```
